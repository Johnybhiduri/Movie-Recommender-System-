# Extracting Only Useful Columns to new_df

```python
new_df = df[['title', 'certificate', 'genre', 'about', 'director', 'stars']]
new_df.head()
```

> These columns are enough for use to create a content based recommender system for our website

```python
new_df.isna().sum()
```

```python
new_df = new_df.dropna().reset_index(drop = True)
```

# Preprocessing

**Steps**

> We will use genre, about, director, stars in single column as tags.

> Remove Stop words

> convert tags into Vectors.

> we will use this function to stem words in about

```python
new_df.head()
```

#### Lets Extract stars names 

```python
import re
```

```python
new_df['stars']  = new_df['stars'].apply(lambda x : re.sub(r"[()']", '', x))
```

```python
new_df.head()
```

#### Removing Spaces between director and stars names to keep them unique

```python
new_df['director'] = new_df['director'].str.replace(' ', '')
```

```python
new_df['stars'] = new_df['stars'].str.replace(' ', '')
```

```python
new_df.head()
```

# Concatenating all the columns except title into 1 column as tags

```python
new_df['genre'] = new_df['genre'].apply(lambda x : x.split())
new_df['certificate'] = new_df['certificate'].apply(lambda x : x.split())
new_df['about'] = new_df['about'].apply(lambda x : x.split())
new_df['director'] = new_df['director'].apply(lambda x : x.split())
new_df['stars'] = new_df['stars'].apply(lambda x : x.split())
```

```python
new_df.head(2)
```

```python
new_df['Tags'] = new_df['certificate'] + new_df['genre'] + new_df['about'] + new_df['director'] + new_df['stars']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return ' '.join(y)
```

```python
new_df.head()
```

```python
final_df = new_df[['title', 'Tags']]
```

```python
final_df.head()
```

```python
final_df['Tags'] = final_df['Tags'].apply(lambda x : ' '.join(x))
```

```python
final_df.head()
```

```python
final_df['Tags'] = final_df['Tags'].apply(stem)
```

#### Converting to lowercase

```python
final_df['Tags'] = final_df['Tags'].apply(lambda x : x.lower())
```

# Converting Tags into vectors and removing stop Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(final_df['Tags']).toarray() # it will convert sparse matrix to numpy array
```

```python
cv.get_feature_names()
```

# Recommendation using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

# Making Recommendation Function

```python
final_df.iloc[2].title
```

```python
def get_recomm(movie):
  movie_index  = final_df[final_df['title'] == movie].index[0]
  recom = sorted(list(enumerate(similarity[movie_index])), reverse = True, key = lambda x : x[1])[1 : 6]
  for i in recom:
    print(final_df.iloc[i[0]].title)
```

# Lets Get Movies Recommendation

```python
get_recomm('X-Men')
```
